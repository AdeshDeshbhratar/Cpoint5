```python
df1.rename(columns={'Date':'Inventory Receive','Inventory':'Total_Item_Qty'}, inplace = True)
```

```python
df2.rename(columns= {"Total Item Qty":"Total_Item_Qty"},inplace = True)
```

```python
merged_df = pd.concat([df1, df2]).reset_index(drop=True)
```

```python
import datetime
```

```python
merged_df['Inventory Receive'] = pd.to_datetime(merged_df['Inventory Receive'])
```

```python
table = merged_df.pivot_table(index='Item Number', columns='Inventory Receive', values='Total_Item_Qty').fillna(0)
```

```python
table
```

```python
table.info()
```

```python
monthly_totals = table.resample('M', axis=1).sum()
```

```python
monthly_totals
```

```python
monthly_totals.plot(kind='bar', figsize=(10, 6))
plt.title('Total Monthly Inventory Levels')
plt.xlabel('Month')
plt.ylabel('Total Inventory')
plt.xticks(rotation=45)
plt.legend(title='Item Number', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```
